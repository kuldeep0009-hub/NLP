```python
import pandas as pd
```

```python
message=pd.read_csv('/content/train.csv')
```

```python
message
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
import re

corpus=[]
#Ye Python ke re (Regular Expression) module ka substitution function hai.
#Matlab: pattern dhundho aur usko replace kar do kisi aur text se.
#re.sub('[^a-zA-Z]', ' ', ...) → sirf alphabets rakhta hai, baaki sab hata deta hai.
# lower() → sabko lowercase me convert.
# split() → sentence ko words me tod deta hai.
# List comprehension → stopwords (like "the", "is", "and") remove karta hai.
# ' '.join(review) → saare cleaned words ko wapas ek single string me convert karta hai.
# Corpus ban jata hai → NLP me use hone ke liye ready.
for i in range(0,len(message)):
  review=re.sub('[^a-zA-Z]',' ',message['sms'][i])
  review=review.lower()
  review=review.split()
  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
##create bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(2,2))
```

```python
x=cv.fit_transform(corpus).toarray()
```

```python
x
```

```python
cv.vocabulary_
```

```python
## output features
##y=pd.get_dummies(message['label']) use jab krna h jab categorical ko one hot encoding me convert krna ho par humara already encoded h
y=message['label']
```

```python
y
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(x_train,y_train)
```

```python
spam_detect_model
```

```python
y_pred=spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
## crating tf dif model
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
x=tv.fit_transform(corpus).toarray()
```

```python
tv.vocabulary_
```

```python
## train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model=MultinomialNB().fit(x_train,y_train)
```

```python
 ## prediction
y_pred=spam_tfidf_model.predict(x_test)
```

```python
score=accuracy_score(y_test,y_pred)
```
